### Modelo de YOLO Pose

Instalamos las dependencias para YOLO

```python
from ultralytics import YOLO
import cv2
```

Probamos el modelo YOLO Pose 11

```python
model = YOLO('yolo11n-pose.pt') 

vid = cv2.VideoCapture(0)

names = None

while True:
    ret, frame = vid.read()
    flipped_frame = cv2.flip(frame, 1)

    if ret:
        results = model(flipped_frame, show=False)
        annotated_frame = results[0].plot()
        print(results[0].keypoints)
        cv2.imshow("Deteccion de YOLO", annotated_frame)

        # Salir del vdeo cuando presionamos ESC
        if cv2.waitKey(1) & 0xFF == 27 or cv2.getWindowProperty("Deteccion de YOLO", cv2.WND_PROP_VISIBLE) < 1:
            break
    else:
        # El vdeo ya se termin
        break

vid.release()
cv2.destroyAllWindows()
```
